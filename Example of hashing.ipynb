```python
def tampilkan_tabel_hash(tabel_hash):
    for i in range(len(tabel_hash)):
        print(i, end=" ")
        for j in tabel_hash[i]:
            print("-->", end=" ")
            print(j, end=" ")
        print()

tabel_hash = [[] for _ in range(10)]

def Hashing(lokasi_item):
    return lokasi_item % len(tabel_hash)

def insert(tabel_hash, lokasi_item, data_item):
    hash_key = Hashing(lokasi_item)
    tabel_hash[hash_key].append(data_item)

insert(tabel_hash, 10, 'Pisang')
insert(tabel_hash, 25, 'Apel')
insert(tabel_hash, 20, 'Jeruk')
insert(tabel_hash, 9, 'Anggur')
insert(tabel_hash, 21, 'Melon')
insert(tabel_hash, 21, 'Mangga')

tampilkan_tabel_hash(tabel_hash)
```
