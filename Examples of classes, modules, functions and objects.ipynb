```python
class GradeCalculator:
    def __init__(self, nilai):
        self.nilai = nilai

    def hitung_grade(self):
        if self.nilai >= 80 and self.nilai <= 100:
            return 'A'
        elif self.nilai >= 60 and self.nilai <= 79:
            return 'B'
        elif self.nilai >= 40 and self.nilai <= 59:
            return 'C'
        else:
            return 'D'

if __name__ == "__main__":
    nilai_mahasiswa = float(input("Masukkan nilai mahasiswa: "))
    nilai = GradeCalculator(nilai_mahasiswa)
    print(f"Grade yang didapatkan: {nilai.hitung_grade()}")
```
