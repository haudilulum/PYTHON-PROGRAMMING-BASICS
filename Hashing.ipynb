```python
#Percobaan ke-1
def hash(number, table_size):
  return number % table_size

Item = 60
Ukuran_Tabel = 23
print (hash(Item, Ukuran_Tabel))
```

```python
#Percobaan ke-2
def hash_string(a_string, table_size):
  sum = 0
  for pos in range(len(a_string)):
    sum = sum + ord(a_string[pos])
  return sum % table_size

print (hash_string('top', 11))
```
