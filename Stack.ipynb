```python
class Stack:
  def __init__ (self):
    self.items = []
  def is_empty(self):
    return self.items == []
  def push(self, item):
    self.items.append(item)
  def pop (self):
    return self.items.pop()
  def peek(self):
    return self.items[len(self.items)-1]
  def size (self):
    return len(self.items)


s = Stack()
print(s.is_empty())
s.push (42)
s.push('Teknik')
print(s.peek())
s.push(32.8)
s.push (True)
print(s.size())
print(s.is_empty())
s.push(8.4)
s.pop()
s.push('Biomedis')
print(s.size())
```
