```python
#Program ini menghitung dan mencetak gaji per minggu
#Aturan Penggajian:
#Gaji reguler: Apabila bekerja hingga 40 jam maka gaji sesuai dengan tarif yang ditentukan.
#Upah lembur: untuk setiap jam di atas 40, ditambah 1,5 dikali dengan tarif yang ditentukan
rate = float(input("Masukan Gaji Per jam: "))
hours = float(input("Masukan Jumlah jam Kerja: "))
if hours > 40:
  pay = 40 * rate + (hours - 40) * 1.5 * rate
  print ("Total Gaji Rp.", pay)
else:
  pay = hours * rate
  print ("Total Gaji Rp.", pay)
```

```python
#Program ini menghitung dan mencetak yang angka terbesar
#dari ketiga angka yang dimasukan
N1 = int(input("Masukkan angka pertama: "))
N2 = int(input("Masukkan angka kedua: "))
N3 = int(input("Masukkan angka ketiga: "))

if N1 > N2:
    if N1 > N3:
        print(N1, "adalah angka terbesar")
    else:
        print(N3, "adalah angka terbesar")
else:
    if N2 > N3:
        print(N2, "adalah angka terbesar")
    else:
        print(N3, "adalah angka terbesar")
```

```python
# Program untuk menentukan apakah suatu angka positif, nol, atau negatif

# Memasukkan angka
angka = float(input("Masukkan angka: "))

# Memeriksa kondisi dan mencetak hasil
if angka > 0:
    print("Angka adalah bilangan positif")
elif angka == 0:
    print("Angka adalah nol")
else:
    print("Angka adalah bilangan negatif")
```
