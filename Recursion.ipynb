```python
#HAUDIL 'ULUM
#PERCOBAAN 1
def list_sum(num_list):
  if len(num_list) == 1:
    return num_list[0]
  else:
    return num_list[0] + list_sum(num_list[1:])

print(list_sum([2, 5, 4, 5, 2]))
```

```python
#HAUDIL 'ULUM
#PERCOBAAN 2
def faktorial(n):
  if n == 0:
    return 1
  else:
    return n*(faktorial)(n-1)

num = int(input("Masukan angka: "))

if num < 0:
  print("Masukan angka positif")
elif num == 0:
  print("Faktorial 0 = 1")
else:
  print("Faktorial dari", num, "=", faktorial (num))
```

```python
def Cetak_Segitiga(A):
    # Base case: if the list has less than 1 element, return
    if len(A) < 1:
        return

    # Create a new list with length one less than the current list
    temp = [0] * (len(A) - 1)

    # Fill the new list with the sums of adjacent elements
    for i in range(len(A) - 1):
        temp[i] = A[i] + A[i + 1]

    # Recursive call with the new list
    Cetak_Segitiga(temp)

    # Print the current list
    print(A)

# Example usage
A = [1, 2, 3, 4, 5]
Cetak_Segitiga(A)
```
