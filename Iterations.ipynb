```python
#while loop
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
#while loop
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
#while loop
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
#while loop
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i sudah tidak kurang dari 6")
```

```python
#for loop
buah = ["apel", "pisang", "mangga"]
for x in buah:
  print(x)
```

```python
#for loop
buah = ["apel", "durian", "pisang", "jambu", "mangga"]
for x in buah:
  print(x)
  if x == "pisang":
    break
```

```python
#for loop
buah = ["apel", "durian", "pisang", "jambu", "mangga"]
for x in buah:
  if x == "pisang":
    continue
  print(x)
```

```python
#for loop
buah = ["apel", "durian", "pisang", "jambu", "mangga"]
sifat = ["merah", "matang", "manis"]

for x in buah:
  for y in sifat:
    print(x, y)
```
