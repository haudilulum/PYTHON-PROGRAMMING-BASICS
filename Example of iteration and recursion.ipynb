```python
def fibo(n):
  a = 0
  b = 1
  for i in range (n):
    temp = a
    a = b
    b = temp + b
  return a

print("Deret Fibonacci: ")
for c in range(0, 10):
    print(fibo(c), end=" ")
```

```python
def faktorial(n):
    if n == 1:
        return n
    else:
        return n * faktorial(n-1)

num = int(input("Masukan angka : "))
if num < 0:
    print("Masukan angka positif")
elif num == 0:
    print("Faktorial 0 = 1")
else:
    print("Faktorial dari", num, "=", faktorial(num))
```

```python
# Menggunakan for loop
def faktorial(n):
    faktorial = 1
    for i in range(1, n + 1):
        faktorial *= i
    return faktorial

num = int(input("Masukkan angka: "))
if num < 0:
    print("Masukan angka positif")
elif num == 0:
    print("Faktorial 0 = 1")
else:
  print("Faktorial dari", num, "=", faktorial(num))
```

```python
# Menggunakan while loop
def faktorial_while_loop(n):
    faktorial = 1
    i = 1
    while i <= n:
        faktorial *= i
        i += 1
    return faktorial

num = int(input("Masukkan angka: "))
if num < 0:
    print("Masukan angka positif")
elif num == 0:
    print("Faktorial 0 = 1")
else:
  print("Faktorial dari", num, "=", faktorial(num))
```
