```python
class rasional:
    def __init__(self, a=0, b=0):
        self.p = a
        self.q = b

def bandingkan(a: rasional, b: rasional) -> int:
    if (a.p * b.q == a.q * b.p):
        return 0
    if (a.p * b.q > a.q * b.p):
        return 1
    return -1

def binarySearch(array: list, l: int, r: int, x: rasional) -> int:
    if r >= l:
        tengah = l + (r - l) // 2
        if bandingkan(array[tengah], x) == 0:
            return tengah
        if bandingkan(array[tengah], x) > 0:
            return binarySearch(array, l, tengah - 1, x)
        return binarySearch(array, tengah + 1, r, x)
    return -1

# Program utama
array = [rasional(1, 5), rasional(2, 3), rasional(3, 2), rasional(13, 2)]
x = rasional(26, 4)

n = len(array)
print("Kelipatan bilangan rasional ditemukan pada indeks:", (binarySearch(array, 0, n - 1, x)))
```

```python
def count_odd_numbers(arr):
    count = 0
    for number in arr:
        if number % 2 != 0:
            count += 1
    return count

# Contoh penggunaan
array = [11, 22, 32, 45, 57, 66, 70, 85, 99]
jumlah_angka_ganjil = count_odd_numbers(array)
print("Jumlah angka ganjil:", jumlah_angka_ganjil)
```
