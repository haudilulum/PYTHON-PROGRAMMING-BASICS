```python
#percobaan kondisi if
usia = int(input("Masukan usia "))
if (usia >=18):
  print("Cukup Umur")
```

```python
#percobaan if else
nilai = int(input("Masukan nilai = "))
if (nilai > 70):
  print("Selamat Anda Lulus")
else:
  print("Maaf Anda Tidak Lulus")
```

```python
#percobaan if elif
var1 = 2+3j
if (type (var1) == int):
  print("Jenis Variabel Integer")
elif (type (var1) == float):
  print("Jenis Variabel Float")
elif (type (var1) == complex):
  print("TJenis Variabel Complex")
elif (type (var1) == bool):
  print("Jenis Variabel Bool")
elif (type (var1) == str):
  print("Jenis Variabel String")
elif (type (var1) == tuple):
  print("Jenis Variabel Tuple")
elif (type (var1) == dict):
  print("Jenis Variabel Dictionaries")
elif (type (var1) == list):
  print("TJenis Variabel List")
else:
  print("Jenis Variabel Tidak Diketahui")
```

```python
#perulangan While Loop
#dieksekusi ketika kondisi bernilai true
count = 1
while count < 9:
  print ("hitung = ",count) #kondisi bernilai true
  count = count + 1
print ('Selesai')
```

```python
# Perulangan For loop
angka = [1,2,3,4,5]
for x in angka:
  print(x)

buah = ["nanas", "apel", "jeruk"]
for makanan in buah:
  print("Saya suka makan ", makanan)
```
