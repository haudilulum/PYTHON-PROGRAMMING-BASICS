```python
#Percobaan 1 Queue
#Teknik Elektro 07
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        return self.items.pop(0)

    def peek(self):
        if not self.is_empty():
            return self.items[0]
        else:
            return None

    def size(self):
        return len(self.items)

j = Queue()
print(j.is_empty())
j.enqueue('Elektro')
j.enqueue(4)
print(j.peek())
j.enqueue(True)
print(j.size())
print(j.is_empty())
j.enqueue(8.4)
j.dequeue()
j.dequeue()
print(j.size())
```

```python
#Percobaan 2 Deque
#Teknik Elektro 07
class Deque:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return self.items == []

    def add_front(self, item):
        self.items.append(item)

    def add_rear(self, item):
        self.items.insert(0, item)

    def remove_front(self):
        return self.items.pop()

    def remove_rear(self):
        return self.items.pop(0)

    def size(self):
        return len(self.items)

    def rear(self):
        return self.items[0]

    def front(self):
        return self.items[-1]

d = Deque()
print(d.is_empty())
d.add_rear(4)
d.add_rear(42)
d.add_front(9.3)
d.add_front('Elektro')
print(d.front())
print(d.rear())
d.remove_front()
d.remove_rear()
print(d.size())
print(d.is_empty())
d.add_front(5.6)
print(d.size())
```
